```python
# random vector generator
import random
import numpy as np


def random_vector(n):
    return np.array([random.uniform(-100, 100) for _ in range(n)])
```

```python
def simulate_matrix(n):
    vec1=random_vector(n)
    vec2=random_vector(n)
    matrix=np.outer(vec1,vec2)
    # pertube matrix
    matrix+=np.random.normal(0,1,matrix.shape)
    print("vec1: ",vec1)
    print("vec2: ",vec2)
    print("outer product: ")
    print(matrix)
    print("vector_norms: ")
    print(np.linalg.norm(vec1))
    print(np.linalg.norm(vec2))
    print("svd: ")
    print(np.linalg.svd(matrix))
    print("normprod:")
    print(np.linalg.norm(vec1)*np.linalg.norm(vec2))
    
```

```python
simulate_matrix(3)
```

```python
A=np.zeros((30,10), dtype=float)
A[0,0]=7
```

```python
from matplotlib import pyplot as plt


x_list=[]
largest_sv=[]
sec_largest_sv=[]
for x in range(0,9):
    x/=10
    print("x:", x)
    B= A+x
    [U,S,VT]=np.linalg.svd(B)
    print("S:", S)
    x_list.append(x)
    largest_sv.append(S[0])
    sec_largest_sv.append(S[1])

# plot the singular values 
plt.plot(x_list,largest_sv, label="largest singular value")
plt.plot(x_list,sec_largest_sv, label="second largest singular value")
```
